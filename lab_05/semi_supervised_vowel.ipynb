```python
# 下载vowel数据集
X, Y = fetch_openml('vowel', return_X_y=True)
```

```python
# 将label编码0-10
le = preprocessing.LabelEncoder()
le.fit(Y)
Y = le.transform(Y)
```

```python
# 构造训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```
